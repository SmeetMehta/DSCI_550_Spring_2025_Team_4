```python
import pandas as pd

# Load the JSON file into a DataFrame
df = pd.read_json(r'haunted_data.json')  # Replace with your actual file name

# Display the DataFrame
df
```

```python
import pandas as pd

# Convert latitude and longitude to numeric (float), coerce errors to NaN
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

# Drop rows where latitude or longitude is missing
df = df.dropna(subset=['latitude', 'longitude'])

summary_df = df.groupby(['city', 'state_abbrev'], as_index=False).agg(
    sighting_counts=('city', 'count'),
    latitude=('latitude', 'mean'),
    longitude=('longitude', 'mean')
)
summary_df = summary_df[['city', 'sighting_counts', 'state_abbrev', 'latitude', 'longitude']]
# summary_df[summary_df['city'].str.lower() == 'marquette']
summary_df
```

```python
# Save the DataFrame as a JSON file
summary_df.to_json("sightings.json", orient="records", indent=4)
```

```python
import pandas as pd

# Load the JSON file into a DataFrame
event_df = pd.read_json(r'D:\viterbi syllabus\dsci550\hw3\haunted_data.json')  # Replace with your actual file name

# Display the DataFrame
event_df = event_df.groupby(['state_abbrev', 'state', 'Event Type']).size().reset_index(name='event_count')
# Drop rows where event_type is 'Unknown'
event_df = event_df[event_df['Event Type'].str.lower() != 'unknown']

```

```python
event_df.to_json("event_type.json", orient="records", indent=4)
```

```python
import json
from collections import defaultdict

# Load your file
with open("event_type.json") as f:
    data = json.load(f)

# Build hierarchy
states = defaultdict(list)
for row in data:
    states[row["state_abbrev"].upper()].append({
        "name": row["Event Type"],
        "value": row["event_count"]
    })

# Final structure
packed_data = {
    "name": "US Events",
    "children": [{"name": state, "children": events} for state, events in states.items()]
}

# Save to new JSON
with open("circle_packing_data.json", "w") as f:
    json.dump(packed_data, f, indent=2)
```

```python
import pandas as pd
import json

# Load the JSON file
witness = pd.read_json(r'haunted_data.json')  # Replace with your actual file name



# Ensure Estimated Eyewitnesses is numeric
witness['Estimated Eyewitnesses'] = pd.to_numeric(witness['Estimated Eyewitnesses'], errors='coerce')

# Drop rows with NaN in essential columns
witness = witness.dropna(subset=['state', 'state_abbrev', 'Estimated Eyewitnesses', 'Caption'])

# Group by state and state_abbrev to get total eyewitnesses
sum_df = witness.groupby(['state', 'state_abbrev'], as_index=False)['Estimated Eyewitnesses'].sum()

# Get caption with the max eyewitnesses per state
idx = witness.groupby(['state', 'state_abbrev'])['Estimated Eyewitnesses'].idxmax()
caption_df = witness.loc[idx, ['state', 'state_abbrev', 'Caption']]

# Merge totals with max-caption
result_df = pd.merge(sum_df, caption_df, on=['state', 'state_abbrev'])

result_df
```

```python
result_df.to_json("witness.json", orient="records", indent=4)
```
